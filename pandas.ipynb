```python
#91.	Chain multiple functions using pipe() to normalize and rank data.
import pandas as pd 
df=pd.read_csv('personality_dataset.csv')
def normalise(df):
    numerical =df.select_dtypes(include='number').columns
    df[numerical] =(df[numerical]-df[numerical].min())/(df[numerical].max()-df[numerical].min())
    return df
def add_rank(df):
    numerical =df.select_dtypes(include='number').columns
    rank =df[numerical].rank(axis=1)
    return pd.concat([rank,df[['Friends_circle_size']]],axis=1)
result =(df.pipe(normalise).pipe(add_rank))
print(result.head())

```
